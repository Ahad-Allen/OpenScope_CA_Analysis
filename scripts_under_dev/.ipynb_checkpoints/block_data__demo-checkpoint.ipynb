<h1> Function Definitions </h1>

```python
def make_behav_sel_df(df, stimtype, mouse_df, mouse_df_fnm, layers, prepost_sec, block_seq_ns, 
                      mouse_n, sess_n, stand, norm, dff=False):

    """
    Main function I'm running to produce dataframes
    """

#     t = time.time()

    layers = {'L23-Cux2': 'L2/3', 'L5-Rbp4': 'L5'}

    idx=df.shape[0]
    print('Mouse ', mouse_n, ', sess ', sess_n)
    sessid = mouse_df[(mouse_df['mouse_n']==mouse_n) & 
                      (mouse_df['sess_n']==sess_n)]['sessid'].values[0]
#     sess = session.Session(datadir, sessid, only_matched_rois=only_matched_rois)
    sess = session.Session(datadir, sessid)
    sess.extract_sess_attribs(mouse_df=mouse_df_fnm)
    sess.extract_info(fulldict=False, roi=True, run=True, pupil=True) 
    stim = sess.get_stim(stimtype)
    layer = layers[mouse_df[(mouse_df['mouse_n']==mouse_n) & 
                            (mouse_df['sess_n']==sess_n)]['line'].values[0]]
    cmptmt = mouse_df[(mouse_df['mouse_n']==mouse_n) & 
                      (mouse_df['sess_n']==sess_n)]['plane'].values[0]

    surp_dict={'gabors':0, 'bricks':1}
    seg_ns = stim.get_segs_by_criteria(gabfr=3, surp=surp_dict[stimtype], remconsec=True, by='seg')

    for i,block_seq_n in enumerate(block_seq_ns['nonsurp']):
        if i==0:
            # roi x seq x time or seq x time
            nonsurp__run_data, nonsurp__pd_data = \
                get_data_tensors(sess, stim, seg_ns, block_seq_n, surp_dict[stimtype], 
                                 prepost_sec[stimtype]['nonsurp']['pre'], 
                                 prepost_sec[stimtype]['nonsurp']['post'], stand, norm, 
                                 reverse=True, dff=dff)
        else:
            # roi x seq x time or seq x time
            nonsurp__run_data__block_seq, nonsurp__pd_data__block_seq = \
                get_data_tensors(sess, stim, seg_ns, block_seq_n, surp_dict[stimtype], 
                                 prepost_sec[stimtype]['nonsurp']['pre'], 
                                 prepost_sec[stimtype]['nonsurp']['post'], stand, norm, 
                                 reverse=True, dff=dff)
            nonsurp__run_data = np.concatenate((nonsurp__run_data, nonsurp__run_data__block_seq), axis=1)
            nonsurp__pd_data  = np.concatenate((nonsurp__pd_data, nonsurp__pd_data__block_seq), axis=1)

    surp=1
    seg_ns = stim.get_segs_by_criteria(gabfr=3, surp=surp, remconsec=True, by='seg')

    for i,block_seq_n in enumerate(block_seq_ns['surp']):
        if i==0:
            # roi x seq x time or seq x time
            surp__run_data, surp__pd_data = \
                get_data_tensors(sess, stim, seg_ns, block_seq_n, surp, 
                                 prepost_sec[stimtype]['surp']['pre'], 
                                 prepost_sec[stimtype]['surp']['post'], stand, norm, 
                                 reverse=False, dff=dff)
        else:
            # roi x seq x time or seq x time
            surp__run_data__block_seq, surp__pd_data__block_seq = \
                    get_data_tensors(sess, stim, seg_ns, block_seq_n, surp, 
                                     prepost_sec[stimtype]['surp']['pre'], 
                                     prepost_sec[stimtype]['surp']['post'], stand, norm, 
                                     reverse=False, dff=dff)
            surp__run_data = np.concatenate((surp__run_data, surp__run_data__block_seq), axis=1)
            surp__pd_data  = np.concatenate((surp__pd_data, surp__pd_data__block_seq), axis=1)

    if stimtype=='bricks' or block_seq_ns['nonsurp'][0]!='all':
        print('not gabor all')
        surp__run_data = np.nanmean(surp__run_data, axis=1)
        nonsurp__run_data = np.nanmean(nonsurp__run_data, axis=1)
        surp__pd_data = np.nanmean(surp__pd_data, axis=1)
        nonsurp__pd_data = np.nanmean(nonsurp__pd_data, axis=1)
    
    run_surp_sel__diff  = surp__run_data - nonsurp__run_data
    pd_surp_sel__diff   = surp__pd_data - nonsurp__pd_data
    
    df.loc[idx] = [mouse_n, layer, cmptmt, stimtype, sess_n,
                   surp__run_data, nonsurp__run_data, run_surp_sel__diff,
                   surp__pd_data, nonsurp__pd_data, pd_surp_sel__diff]
    
    return df
```

```python
def get_twop_stim_ns_df(sess, gab_seg_ns, surp, pre, post, datatype):
    """
    Get twop or stim segment ns and associated info for Gabors
    This is working for one frame (e.g., "U"). Test for multiple!
    """

#     surp = 1
#     gab_seg_ns = sess.gabors.get_segs_by_criteria(gabk=16, gabfr=3, surp=surp, by='seg')
#     pre = 0
#     post = 0.6
    # data_ = sess.gabors.get_twop_fr_by_seg(gab_seg_ns, first=True, ch_fl=[pre, post])['first_twop_fr']
    # stim_fr_ns = sess.gabors.get_stim_fr_by_seg(gab_seg_ns, first=True, ch_fl=[pre, post])['first_stim_fr']
    # pup_fr_ns = sess.get_pup_fr_by_twop_fr(data_.tolist(), ch_fl=[pre, post])

    gab_block_ns = []
    block_seq_ns = []
    block_n_seqs = []

    if datatype=='twop':
        twop_stim_ns_df = sess.gabors.get_twop_fr_by_seg(gab_seg_ns, first=True, ch_fl=[pre, post])
    elif datatype=='stim':
        twop_stim_ns_df = sess.gabors.get_stim_fr_by_seg(gab_seg_ns, first=True)#, ch_fl=[pre, post])

    twop_stim_ns_df.index = twop_stim_ns_df.index.rename('global_seq_n')
    twop_stim_ns_df['gab_seg_n'] = gab_seg_ns

    gab_block_first_seg_ns = np.array(sess.gabors.get_segs_by_criteria(gabk=16, gabfr='any', 
                                                                       surp=surp, by='seg', remconsec=True))
    for seg_n in twop_stim_ns_df['gab_seg_n']:
        gab_block_n = np.max(np.where(gab_block_first_seg_ns <= seg_n))
        block_seq_n = seg_n - gab_block_first_seg_ns[gab_block_n]
        block_seq_n = np.floor_divide(seg_n - gab_block_first_seg_ns[gab_block_n], 4)
        gab_block_ns.append(gab_block_n)
        block_seq_ns.append(block_seq_n)
    twop_stim_ns_df['gab_block_n'] = gab_block_ns
    twop_stim_ns_df['block_seq_n'] = block_seq_ns
    for gab_block_n in twop_stim_ns_df['gab_block_n']:
        mask = twop_stim_ns_df['gab_block_n']==gab_block_n
        block_n_seqs.append(twop_stim_ns_df[mask].shape[0])
    twop_stim_ns_df['block_n_seqs'] = block_n_seqs
    twop_stim_ns_df['block_seq_n_reverse'] = twop_stim_ns_df['block_n_seqs'] - \
        twop_stim_ns_df['block_seq_n'] - 1
    return twop_stim_ns_df
```

```python
def get_data_tensors(sess, stim, seg_ns, block_seq_n, surp, pre, post, stand, norm,
                     reverse=False, dff=False, full_pd=False):
    """
    Get data dfs from block sequence number(s)
    Test with list of ns
    """
    
    if reverse==False:
        block_seq_str = 'block_seq_n'
    else:
        block_seq_str = 'block_seq_n_reverse'
    
    if stimtype=='bricks':  
        twop_fr_ns = stim.get_twop_fr_by_seg(seg_ns, first=True, 
                                             ch_fl=[pre, post])['first_twop_fr']
        stim_fr_ns = stim.get_stim_fr_by_seg(seg_ns, first=True)['first_stim_fr']#, ch_fl=[pre, post])
    elif block_seq_n!='all':
        twop_df    =  get_twop_stim_ns_df(sess, seg_ns, surp, pre, post, 'twop')
        stim_df    =  get_twop_stim_ns_df(sess, seg_ns, surp, pre, post, 'stim')
        mask0      =  twop_df[block_seq_str]==block_seq_n
        twop_fr_ns =  twop_df[mask0]['first_twop_fr']
        mask0      =  stim_df[block_seq_str]==block_seq_n
        stim_fr_ns =  stim_df[mask0]['first_stim_fr']
        if surp:
            print('\n\n\nUnexpected Gabor 2p data frame:\n\n')
        else:
            print('\n\n\nExpected Gabor 2p data frame:\n\n')
        display(twop_df.head())
        print('\n\n')

    # weirdly, errors thrown for reverse if not .to_list(), not for non-reverse (??)
    if block_seq_n=='all' and stimtype=='gabors':  #in this case, we have to average over each block
        twop_df =  get_twop_stim_ns_df(sess, seg_ns, surp, pre, post, 'twop')
        stim_df =  get_twop_stim_ns_df(sess, seg_ns, surp, pre, post, 'stim')
        if dff:
            dff_data = get_gab_block_data(stim, sess, seg_ns, twop_df, pre, post, 
                                          mouse_n, sess_n, surp, 'dff')
        print('\nrun data\n')
        run_data = get_gab_block_data(stim, sess, seg_ns, stim_df, pre, post, 
                                      mouse_n, sess_n, surp, 'run')
        print('\npd data\n')
        pd_data  = get_gab_block_data(stim, sess, seg_ns, twop_df, pre, post, 
                                      mouse_n, sess_n, surp, 'pd')
#         pd_data = run_data

    else:
        if dff:
            roi_data_df = stim.get_roi_data(twop_fr_ns.to_list(), pre, post, scale=stand['dff'])
            # roi x seq x time
            dff_data = gen_util.reshape_df_data(roi_data_df, squeeze_rows=False, squeeze_cols=True)
        run_data_df = stim.get_run_data(stim_fr_ns.to_list(), pre, post, scale=stand['run'])
        pup_fr_ns = sess.get_pup_fr_by_twop_fr(twop_fr_ns.tolist(), ch_fl=[pre, post])
        pd_data_df  = stim.get_pup_diam_data(pup_fr_ns, pre, post, scale=stand['pd'])
        # seq x time
        run_data = gen_util.reshape_df_data(run_data_df, squeeze_rows=False, squeeze_cols=True)
        pd_data  = gen_util.reshape_df_data(pd_data_df, squeeze_rows=False, squeeze_cols=True)
    
    if norm['run']:
        full_run_data_df = stim.sess.get_run_velocity(scale=stand['run'])
    if norm['pd']:
        full_pd_data_df = stim.sess.get_pup_data(scale=stand['pd'])

    if norm['run']:
        max_run_vel = np.nanmax(full_run_data_df.to_numpy())
        min_run_vel = np.nanmin(full_run_data_df.to_numpy())
        run_vel_range = max_run_vel - min_run_vel
        run_data = (run_data - min_run_vel) / (run_vel_range)
    if norm['pd']:
        max_pd = np.nanmax(full_pd_data_df.to_numpy())
        min_pd = np.nanmin(full_pd_data_df.to_numpy())
        pd_range = max_pd - min_pd
        pd_data = (pd_data - min_pd) / (pd_range)
    
    if dff:
        print(dff_data.shape, run_data.shape, pd_data.shape)
    else:
        print(run_data.shape, pd_data.shape)

    if dff:
        return dff_data, run_data, pd_data
    else:
        return run_data, pd_data
```

```python
def get_gab_block_data(stim, sess, seg_ns, df, pre, post, mouse_n, sess_n, surp, data_type):

    fr_type = {'dff':'first_twop_fr', 'run':'first_stim_fr', 'pd':'first_twop_fr'}

#     mouse_sess_blk_stim_surp_exclusions = get_exclusions()
#     print('mouse_sess_blk_stim_surp_exclusions', mouse_sess_blk_stim_surp_exclusions)
    
#     mouse_sess_stim_surp_exclusions = []
#     mouse_sess_stim_surp_exclusions.append('{:02}{}{}{}'.format(1, 2, 'pd', 0))
#     mouse_sess_stim_surp_exclusions = np.array(mouse_sess_stim_surp_exclusions)
#     print('mouse_sess_stim_surp_exclusions', mouse_sess_stim_surp_exclusions)

    data = []
    print('number of blocks: ', df['gab_block_n'].max())
    for block in range(df['gab_block_n'].max()):
        mouse_sess_blk_stim_surp = '{:02},{},{:02},{},{}'.format(mouse_n, sess_n, block, 
                                                             data_type, surp)
#         mouse_sess_stim_surp = '{:02}{}{}{}'.format(mouse_n, sess_n, block, 
#                                                              data_type, surp)
#         if not np.isin(mouse_sess_blk_stim_surp, mouse_sess_blk_stim_surp_exclusions).any():
        print('block ', block, 'mouse_sess_blk_stim_surp ', mouse_sess_blk_stim_surp)
        mask0  =  df['gab_block_n']==block
        fr_ns =  df[mask0][fr_type[data_type]]
        if data_type=='dff':
            data_df = stim.get_roi_data(fr_ns.to_list(), pre, post, scale=stand['dff'])
        elif data_type=='run':
            data_df = stim.get_run_data(fr_ns.to_list(), pre, post, scale=stand['run'])
        elif data_type=='pd':
#             import pdb
#             pdb.set_trace()
            pup_fr_ns = sess.get_pup_fr_by_twop_fr(fr_ns.tolist(), ch_fl=[pre, post])
#                 pup_fr_ns = sess.get_pup_fr_by_twop_fr(fr_ns.tolist())
            data_df  = stim.get_pup_diam_data(pup_fr_ns, pre, post, scale=stand['pd'])
        blk_data = gen_util.reshape_df_data(data_df, squeeze_rows=False, squeeze_cols=True)
        #average over time:
        blk_data = np.nanmean(blk_data, axis=-1)
        #average over trials in the block:
        blk_data = np.nanmean(blk_data, axis=-1)
        data.append(blk_data)
        
        
    return np.array(data)
```
